```python
import tensorflow as tf
from sklearn.preprocessing import StandardScaler
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('./XAUUSD_historical_data.csv')
data = data.iloc[::-1].reset_index(drop=True)
data["Target"] = data["Close"].shift(-1)
data = data.dropna()
data["Change(Pips)"] = data["Change(Pips)"]/100
X = data.drop(columns = ["Date", "Target"])
y = data["Target"]
data
```

```python
sc_X = StandardScaler()
sc_y = StandardScaler()

X_transformed = sc_X.fit_transform(np.array(X))
y_transformed = sc_y.fit_transform(np.array(y).reshape(-1,1))
```

```python
from tensorflow.keras.models import load_model
loaded_model = load_model('./predictor_model.keras')
```

```python
candle_open = input("Open: ")
candle_high = input("High: ")
candle_low = input("Low: ")
candle_close = input("Close: ")
candle_change = input("Change: ")
candle_change_percentage = input("Change Percentage: ")

print(sc_y.inverse_transform(loaded_model.predict(sc_X.transform([[candle_open,candle_high,candle_low,candle_close,candle_change,candle_change_percentage]])))[0])
```
